```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# settings to display all columns
pd.set_option("display.max_columns", None)

# settings to display five rows
pd.set_option("display.max_rows", 5)
```

```python
# read post quit random EMAs from the csv file
post_quit_random_df = pd.read_csv('PNS_EMA_data_for_end_users\post_quit_random_ema.csv', sep=',')
post_quit_random_df
```

```python
# read pre quit random EMAs from the csv file
pre_quit_random_df = pd.read_csv('PNS_EMA_data_for_end_users\pre_quit_random_ema.csv', sep=',')
pre_quit_random_df.head()
```

```python
#concatenate dataframes
frames = [pre_quit_random_df, post_quit_random_df]
df = pd.concat(frames, sort=False)

# reset index
df.reset_index(drop=True, inplace=True)
df
```

```python
df.insert(5,'study_day','')
```

```python
# calculate the study day from the EMA delivered time and smoking quit time
df['delivered_hrts'] = pd.to_datetime(df['delivered_hrts'])
df['quit_hrts'] = pd.to_datetime(df['quit_hrts'])
#df['study_day'] = (df['delivered_hrts']-df['quit_hrts']).dt.days
df['study_day'] = (df['delivered_hrts'].dt.date-df['quit_hrts'].dt.date).dt.days
df
```

```python
# rename affects with their respective values
df = df.rename(columns={'Affect1': 'Enthusiastic', 'Affect2': 'happy', 'Affect3':'relaxed',
                        'Affect4':'bored','Affect5':'sad', 'Affect6':'angry', 'Affect7':'anxious' , 'Affect8':'restless', 
                        'Affect9':'stressed','Affect10':'hostile'})
df
```

```python
df.study_day.max()
```

```python
df.to_csv(r'PNS_EMA_data_for_end_users\curated_random_ema_pns_data.csv', index=False)
```

### Filter based on participants with atleast one EMA per day

```python
tvem_df = df[(df['study_day'] >= -7) & (df['study_day'] <= 20)]
tvem_df
```

```python
tvem_df.to_csv(r'PNS_EMA_data_for_end_users\curated_random_ema_pns_data.csv', index=False)
```

#### Run below cell ONLY if we want to eliminate first and last day while caluclating active users

```python
test_df = tvem_df.groupby(["id","study_day"]).apply(lambda x: x[x["study_day"].between(-6, 20, inclusive = True)]["delivered_hrts"].count()).reset_index(name='ema_count')
test_df['ema_indicator'] = 0
test_df.loc[test_df['ema_count']>=1, 'ema_indicator'] = 1
df2 = test_df.groupby('id').filter(lambda g: g['ema_indicator'].sum() == 27)
filtered_part_id_list = df2.id.unique()
tvem_df_1 = tvem_df[tvem_df['id'].isin(filtered_part_id_list)]
tvem_df_1
```
