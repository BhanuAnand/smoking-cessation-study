# Missing Data Analysis

```python
df.groupby('study_day').apply(lambda x: x[x['record_status'] == 'Completed']['record_id'].count())
```

```python
day_part_count_df = df.groupby('study_day')['id'].nunique().reset_index()

comp_df = df.groupby('study_day').apply(lambda x: x[x['record_status'] == 'Completed']['record_id'].count()).reset_index()
comp_df.columns = ['study_day', 'complete_ema_count']

incomp_df = df.groupby('study_day').apply(lambda x: x[(x['record_status'] == 'FRAGMENT RECORD') | ((x['record_status'] == 'Incomplete/Timed Out') & (x['with_any_response'] == 1))]['record_id'].count()).reset_index()
incomp_df.columns = ['study_day', 'incomplete_ema_count']

missing_df = df.groupby('study_day').apply(lambda x: x[(x['with_any_response'] == 0)]['record_id'].count()).reset_index()
missing_df.columns = ['study_day', 'missing_ema_count']

total_df = df.groupby('study_day').agg({'record_id': ['count']}).reset_index()
total_df.columns = ['study_day', 'total_ema_count']

total_comp_df = total_df.merge(comp_df, how = 'outer')
total_comp_df['complete_ema_percent'] = round((total_comp_df['complete_ema_count'] / total_comp_df['total_ema_count']) * 100, 2)

total_comp_miss_df = total_comp_df.merge(missing_df, how = 'outer')
total_comp_miss_df['missing_ema_percent'] = round((total_comp_miss_df['missing_ema_count'] / total_comp_miss_df['total_ema_count']) * 100, 2)

total_comp_miss_incomp_df = total_comp_miss_df.merge(incomp_df, how = 'outer')
total_comp_miss_incomp_df['incomplete_ema_percent'] = round((total_comp_miss_incomp_df['incomplete_ema_count'] / total_comp_miss_incomp_df['total_ema_count']) * 100, 2)

total_comp_miss_incomp_df['total_percent'] = total_comp_miss_incomp_df['complete_ema_percent'] + total_comp_miss_incomp_df['missing_ema_percent'] + total_comp_miss_incomp_df['incomplete_ema_percent']

total_new_df = total_comp_miss_incomp_df.merge(day_part_count_df, how = 'outer')

total_new_df['complete_avg_ema'] = round((total_comp_miss_incomp_df['complete_ema_count'] / total_new_df['id']), 2)

total_new_df['incomplete_avg_ema'] = round((total_comp_miss_incomp_df['incomplete_ema_count'] / total_new_df['id']), 2)

total_new_df['missing_avg_ema'] = round((total_comp_miss_incomp_df['missing_ema_count'] / total_new_df['id']), 2)

total_new_df
```

```python
total_comp_miss_incomp_df['new_day_modified'] = total_comp_miss_incomp_df['study_day']+7
```

```python
# create data
t = total_comp_miss_incomp_df.shape[0]
i = 0
x = []
y1_list = []
y2_list = []
y3_list = []

for i in range(0,t):
    
    x.append(str(round(total_comp_miss_incomp_df.study_day[i])))
    
    y1_list.append(total_comp_miss_incomp_df.complete_ema_percent[i])
    y2_list.append(total_comp_miss_incomp_df.incomplete_ema_percent[i])
    y3_list.append(total_comp_miss_incomp_df.missing_ema_percent[i])
    
        
y1 = np.array(y1_list)
y2 = np.array(y2_list)
y3 = np.array(y3_list)
```

```python
# plot bars in stack manner
# Choose columns in the order to "stack" them
plt.figure(figsize=(9, 6))
plt.title("Average number complete, incomplete and missing random EMAs during the study in PNS")
plt.xlabel("Day in the study")
plt.ylabel("Average number of EMAs per day")

plt.bar(x, y1, color='green', label="Complete")
plt.bar(x, y2, bottom=y1, color='yellow', label="Incomplete")
plt.bar(x, y3, bottom=y1+y2, color='red', label="Missing")
plt.legend(bbox_to_anchor =(1.00, 1.00), ncol = 1)

plt.show()
```

```python
# calculate day_time which is to be used for TVEM analysis
def cal_day_time(row):
    day_time = row['study_day'] + round((row['delivered_hrts'].hour/24), 3)
    return day_time

df.insert(6,'day_time','')
df['day_time'] = df.apply(cal_day_time, axis=1)
```

```python
# calculate time_of_day which is to be used for TVEM analysis
def cal_time_of_day(row):
    time = round((row['delivered_hrts'].hour/24), 3)
    return time

df.insert(7,'time_of_day','')
df['time_of_day'] = df.apply(cal_time_of_day, axis=1)
```

```python
# calculate whether an EMA is completed or not based on column record_status
def cal_y(row):
    if row['record_status'] == 'Completed':
        return 1
    else:
        return 0

df.insert(15,'engagement','')
df['engagement'] = df.apply(cal_y, axis=1)
```

#### Analysis based on averages

```python
new_df = df.groupby(['id','study_day'])[['engagement','Enthusiastic','hostile','happy','relaxed','bored','sad','angry','anxious','restless','stressed','Urge1','Urge2','Urge3']].mean().round(2).reset_index()
new_df
```

```python
new_df.rename(columns = {'engagement':'current_day_avg_engagement'}, inplace = True)
```

```python
new_df.insert(3,'previous_day_avg_engagement','')
new_df['previous_day_avg_engagement'] = new_df.groupby('id')['current_day_avg_engagement'].shift(1)
new_df['previous_day_avg_engagement'] = new_df['previous_day_avg_engagement'].fillna(0)
new_df
```

```python
new_df.insert(4,'next_day_avg_engagement','')
new_df['next_day_avg_engagement'] = new_df.groupby('id')['current_day_avg_engagement'].shift(-1)
new_df['next_day_avg_engagement'] = new_df['next_day_avg_engagement'].fillna(0)
new_df
```

```python
new_df['all_previous_days_avg_engagement'] = round((new_df.groupby(['id'])['current_day_avg_engagement'].cumsum())/(new_df.groupby(['id'])['current_day_avg_engagement'].cumcount()+1),2)
new_df.insert(5,'all_previous_days_avg_engagement_1','')
new_df['all_previous_days_avg_engagement_1'] = new_df.groupby('id')['all_previous_days_avg_engagement'].shift(1)
new_df['all_previous_days_avg_engagement_1'] = new_df['all_previous_days_avg_engagement_1'].fillna(0)
new_df.pop('all_previous_days_avg_engagement')
new_df.rename(columns = {'all_previous_days_avg_engagement_1':'all_previous_days_avg_engagement'}, inplace = True)
new_df
```
